```python
1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.

T(n) = 3T(n-1) + 12n

We know that for n=0, T(0)=5
Therefore we put n-1=0 => n=1

T(1) = 3T(1-1) + 12x1 = 3T(0) + 12 = 3 x 5 + 12 = 27
T(2) = 3T(2-1) + 12x2 = 3T(1) + 24 = 3 x 27 + 24 = 105

T(2) = 105
```
